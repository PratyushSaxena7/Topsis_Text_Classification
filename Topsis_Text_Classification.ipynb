```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
FinalAnswer = pd.read_csv('output.csv')
```

```python
FinalAnswer
```

```python
model_names = FinalAnswer['model']
topsis_scores = FinalAnswer['Topsis Score']

# Plotting the bar graph
plt.figure(figsize=(5,5))
plt.bar(model_names, topsis_scores, color='green')
plt.xlabel('Model Name')
plt.ylabel('Topsis Score')
plt.title('Evaluation of Text Classification models using Topsis')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```
